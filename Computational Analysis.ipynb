```python
import numpy as np
import matplotlib.pyplot as plt
```

#### GBM methods (from the grid search methods)

```python
# hyperparams for GBM method (applies to both 28x28 and 64x64 image datasets)
# the only hyperparam used is the number of iterations/trees
n_iters = 250 * np.arange(1, 13)

# Mean fit times for GBM methods (from the grid search methods)
gbm_28_meanfit = [ 89.53192267, 218.67002072, 264.34419847, 337.99569306,
       425.16791453, 425.06064897, 444.90950327, 524.99958506,
       422.11870403, 398.27469687, 511.14828043, 507.47786431]
gbm_64_meanfit = [ 325.02858615,  561.10428834,  822.57409205, 1096.353338  ,
       1366.35772114, 1727.40891113, 1868.52942638, 1882.11907291,
       2046.22780457, 2064.03722572, 2274.76243029, 2544.24600091]
```

```python
fig = plt.figure()
plt.plot(n_iters, gbm_28_meanfit) # TODO (Tara): include axis labels and titles
fig.suptitle("Mean Fit Time for the GBM method, Grid Search Methods (28X28 images)")
plt.xlabel("Number of Iterations")
plt.ylabel("Mean fit times from grid search methods")
```

```python
fig = plt.figure()
plt.plot(n_iters, gbm_64_meanfit) # TODO (Tara): include axis labels and titles
fig.suptitle("Mean Fit Time for the GBM method, Grid Search Methods(64X64 images)")
plt.xlabel("Number of Iterations")
plt.ylabel("Mean fit times from grid search methods")
```

```python
#together
plt.plot(n_iters, gbm_28_meanfit, label='GBM (28 x 28 images)')
plt.plot(n_iters, gbm_64_meanfit, label='GBM (64 x 64 images)')
fig.suptitle("Mean Fit Time for the GBM method -- Grid Search Methods")
plt.xlabel("Number of Iterations")
plt.ylabel("Mean fit times from grid search methods")
```

```python
#together, Difference
dif = np.subtract(gbm_64_meanfit , gbm_28_meanfit)
plt.plot(n_iters, dif)
fig.suptitle("Mean Fit Time for the GBM method -- Grid Search Methods")
plt.xlabel("Number of Iterations")
plt.ylabel("Difference in Mean Fit Time Between 64X64 and 28X28")
```
